**DATA LOADING**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("INvideos.csv")

# Show first 5 rows
df.head()
```

**DATA ANALYSIS**

```python
# Dataset shape
print("Rows and Columns:")
print(df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

**DATA VISUALIZATION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Top 10 trending channels

top_channels = df['channel_title'].value_counts().head(10)

plt.figure(figsize=(12,6))
top_channels.plot(kind='bar')

plt.title("Top Trending Channels")
plt.xlabel("Channel Name")
plt.ylabel("Trending Count")

plt.show()
```

**Likes vs Views Graph**

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='views',
    y='likes',
    data=df
)

plt.title("Likes vs Views")

plt.show()
```

**Correlation Heatmap**

```python
correlation = df[['views', 'likes', 'dislikes', 'comment_count']].corr()

plt.figure(figsize=(8,6))

sns.heatmap(correlation, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
